Create a function that checks each row in the 311 dataframe to see if the word "graffiti" is present in either the complaint_type value or descriptor value. Both columns may contain the word, so you should check both. If the word "graffiti" is found, the function should return the boolean value True. If "graffiti" is not found, the function should return the boolean value False.

```
# Create a function graffiti that flags if complaint type or descriptor contains graffiti
def graffiti(row):
  if pd.notnull(row.descriptor):
    if 'graffiti' in row.descriptor:
      return 'True'
    elif 'graffiti' in row.complaint_type:
      return 'True'
    else:
      return 'False'

type(graffiti)
```

```
# Apply function that creates new column true/false for whether complaint or descriptor contains graffiti
df['graffiti_flag'] = df.apply(graffiti, axis=1)
df.head()
```

```
# Create a dataframe that groups by graffiti flag
df_2 = df.groupby('graffiti_flag').size().reset_index(name='count')

# Check to confirm that the function worked
df_2
```

```
# Check to make sure that more than one complaint_type is included
df_3 = df[df.graffiti_flag == 'True'].groupby('complaint_type').size().reset_index(name='count')

df_3
```

```
# Create another dataframe df_graffiti that only contains records where graffiti_flag is True 
df_graffiti = df[df.graffiti_flag == 'True']
```
