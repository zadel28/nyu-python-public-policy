Q2. Get the count of each submission method per complaint type, sorted by complaint type. Review the results and share a sentence or two with your observations: can you provide a potential explaination for why some complaints are more frequently made via phone and mobile versus on the website?

***It seems as though the complaints relating to public issues/issues outside the home are most commonly made either by phone or mobile because those are the more accessible means of reaching 311 (e.g. air quality, animal in a park,graffiti, etc.). Whereas complaints about shelter or living conditions (e.g. indoor air quality, indoor sewage, plumbing, etc.) seem to more commonly be made online or by phone, which is likely more accessible when people are at home.***


```
top_complaint_methods = df.groupby(['open_data_channel_type', 'complaint_type']).size().reset_index(name='count')
```

```
top_complaint_methods.sort_values('complaint_type', ascending=True)
```
