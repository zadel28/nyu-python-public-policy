```
df = pd.read_csv('/content/drive/My Drive/Data for Python/cleaned_311_data_revised.csv', header='infer')
```

```
df.head()
```

```
# Step 3. Convert the created_date column to be a datetime series.

df.created_date = pd.to_datetime(df.created_date, format='%m/%d/%Y %I:%M:%S %p')
```

```
df.head()
```

```
df.dtypes
```

```
# Step 1. Create a list of the top 10 complaint types (i.e. the complaint type categories that have the most records in our 311 dataset). Save the list as an object called top_10.

top_10 = df.groupby('complaint_type').size().nlargest(10).reset_index(name='count')
```

```
# Step 2. Filter the 311 dataset to only 311 requests that have complaint types in top_10. Save this subset of data as a dataframe called filtered_311.

filtered_311 = df[df["complaint_type"].isin(["Noise - Residential", "HEAT/HOT WATER", "Illegal Parking", "Request Large Bulky Item Collection", "Blocked Driveway", "Noise - Street/Sidewalk", "Street Condition", "UNSANITARY CONDITION", "Water System", "Noise"])]
filtered_311.head()
```

```
filtered_311.dtypes
```

```
# Step 4. Create 6 new columns in filtered_311 that extract each of the following from created_date: year, date, month, weekday, weekday_name, hour

filtered_311.loc[:,('year')] = filtered_311.created_date.dt.year
filtered_311.loc[:,('date')] = filtered_311.created_date.dt.date
filtered_311.loc[:,('month')] = filtered_311.created_date.dt.month
filtered_311.loc[:,('weekday')] = filtered_311.created_date.dt.weekday
filtered_311.loc[:,('weekday_name')] = filtered_311.created_date.dt.weekday_name
filtered_311.loc[:,('hour')] = filtered_311.created_date.dt.hour

filtered_311.head()
```

```
# Find the count of 311 requests per hour and date for each complaint type. Hint: Use the pandas groupby function to aggregate the filtered_311 dataframe by hour, date, and complaint_type and count the number of 311 requests. Save the result as a new dataframe.

complaints_per_date_hour = filtered_311.groupby(['date', 'hour', 'complaint_type']).size().reset_index(name='count_requests')
complaints_per_date_hour.head()
```

```
# Find the median count of 311 requests per hour for each complaint type

complaints_per_hour = complaints_per_date_hour.groupby(['hour', 'complaint_type'])['count_requests'].median().reset_index(name='median_requests')
complaints_per_hour.head()
```

```
# Produce a line plot that has 10 lines on it, one for each of the top 10 complaint types. The x values will be hours and the y values will be the median count of requests per complaint type. Use a for loop to produce the 10 lines all one one figure.
plt.figure(figsize=(20, 6))

for i in top_10.complaint_type:
  df2 = complaints_per_hour[complaints_per_hour.complaint_type == i]
  
  plt.plot(df2.hour, df2.median_requests, label = i)

plt.xticks(df2.hour)
plt.grid()
plt.title('Count of Complaints per Hour, By Complaint Type')
plt.xlabel('Hour')
plt.ylabel('Median Count of 311 Requests')
plt.legend(loc=2)
```

```
# Step 6. Trends by day of week

complaints_per_weekday = filtered_311.groupby(['weekday', 'date', 'complaint_type']).size().reset_index(name='count_requests')
complaints_per_weekday.head()
```

```
complaints_per_week = complaints_per_weekday.groupby(['weekday', 'complaint_type'])['count_requests'].median().reset_index(name='median_requests')
complaints_per_week.head()
```

```
plt.figure(figsize=(20, 6))

for i in top_10.complaint_type:
  df3 = complaints_per_week[complaints_per_week.complaint_type == i]
  
  plt.plot(df3.weekday, df3.median_requests, label = i)

plt.xticks(df3.weekday)
plt.grid()
plt.title('Count of Complaints per Weekday, By Complaint Type')
plt.xlabel('Weekday')
plt.ylabel('Median Count of 311 Requests')
plt.legend(loc=2)
```

```
# Step 7. Trends by month. aggregate by month, year (not date), and complaint_type. Produce a similar plot where x values will month and y values will be the median count of requests per complaint type

complaints_per_month = filtered_311.groupby(['month', 'year', 'complaint_type']).size().reset_index(name='count_requests')
complaints_per_month.head()
```

```
complaints_month = complaints_per_month.groupby(['month', 'complaint_type'])['count_requests'].median().reset_index(name='median_requests')
complaints_month.head()
```

```
plt.figure(figsize=(20, 6))

for i in top_10.complaint_type:
  df4 = complaints_month[complaints_month.complaint_type == i]
  plt.plot(df4.month, df4.median_requests, label = i)

plt.xticks(df4.month)
plt.grid()
plt.title('Count of Complaints per Month, By Complaint Type')
plt.xlabel('Month')
plt.ylabel('Median Count of 311 Requests')
plt.legend(loc=2)
```
